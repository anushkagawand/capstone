```python
import pandas as pd
import numpy as np # library to handle data in a vectorized manner
```

```python
citydf = pd.read_csv('city-data.csv', index_col = 'CityState')
```

```python
myplot1 = citydf.plot(y=["Population"], kind='bar', figsize=(10,3));
myplot1.set_title("Citywise Population", fontsize=30)
myplot1.set_xlabel('CityState', fontsize=18)
myplot1.set_ylabel('Population', fontsize=18);
```

```python
myplot2 = citydf.plot(y=["NosOfGym"], kind='bar', figsize=(10,3));
myplot2.set_title("Number of Gym / Fitness Centers", fontsize=30)
myplot2.set_xlabel('CityState', fontsize=18)
myplot2.set_ylabel('Number of Gym', fontsize=18);
```
